# Data Analysis Intro

## Importing Data

```python
import pandas as pd
```

```python
names = ['id', 'title', 'year', 'rating', 'votes', 'length', 'genres']
data = pd.read_csv('ipynb-data-files/imdb_top_10000.txt', sep="\t", names=names, index_col=0)
```

## Plotting

```python
%matplotlib inline
```

```python
data.plot()
```

```python
data.plot(kind='scatter', x='rating', y='votes')
```

```python
data.plot(kind='scatter', x='rating', y='votes', alpha=0.3)
```

```python
data['rating'].plot(kind='hist')
```

# Data Analysis Intro

## Import Pandas

```python
import pandas as pd
```

## Importing Data

```python
names = ['id', 'title', 'year', 'rating', 'votes', 'length', 'genres']
data = pd.read_csv('ipynb-data-files/imdb_top_10000.txt', sep="\t", names=names, index_col=0)
```

## Advanced Data Selection

```python
data[data['year'] > 1995]
```

```python
data[data['year'] == 1966]
```

```python
data[(data['year'] > 1955) & (data['year'] < 2000)]
```

```python
data[(data['year'] > 1955) & (data['year'] < 2000)].sort_values(by='rating', ascending=False).head(10)
```
